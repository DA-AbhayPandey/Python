```python
df1.merge(df2) # merge() command merge 2 table with comman elements default i.e FellowshipID and FirstName
```

```python
 df1.merge(df2, how = 'inner') # how command in merge() used for describe inner, outer etc
```

```python
 df1.merge(df2, how = 'inner', on = 'FellowshipID') # on = command used to describe what element we want to use as common
                                                    # it only use give 'FellowshipID' as a common and show Name as different of 
                                                    #both table because it not commonn as per command
```

```python
 df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName']) # now it used fellowshipId & FirstName as common 
                                                                  #element as per command
```

```python
 df1.merge(df2, how = 'outer')
```

```python
 df1.merge(df2, how = 'right', on = 'FellowshipID')
```

```python
 df1.merge(df2, how = 'left', on = 'FellowshipID')
```

```python
 df1.merge(df2, how = 'cross') # how=cross command is like a matrix in maths,
                                #it compare all elements from table1 with all element in Table2
```

```python
df1.join(df2, on ='FellowshipID', how = 'outer', lsuffix='left', rsuffix ='right') #join not works as df1.join(df2) because it can't define comman element as default
                                                                                  # we need to use all on =, how=,lsuffix=,rsuffix command   
```

```python
df4= df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),how ='outer', lsuffix='left', rsuffix ='right')
df4                  #join work on index and if we set same index for both table then it works normally
```

```python
pd.concat([df1,df2]) #pd.concat() work as a paste, it paste table2 below table1
```

```python
pd.concat([df1,df2], join='inner') # join ='inner' in pd.concat shows columns which are same  in both table 
```
