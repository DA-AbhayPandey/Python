```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\ASUS\Downloads\Data Analyst\Python\Python File sorter\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#df['Last_Name']=df['Last_Name'].str.lstrip('...')
#df['Last_Name']=df['Last_Name'].str.lstrip('/')
#df['Last_Name']=df['Last_Name'].str.rstrip('_')
df['Last_Name']=df['Last_Name'].str.strip('123._/') #to remove special character from column in one go
                                                    #give command 123 accordingly how many character you want to remove
                                                    # 7 right after enter special character
df
```

```python
#df[["Address","State","Zip Code"]]=df["Address"].str.split(',', expand=True)
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.replace('N/a','') # str.replace used for remove elements from 1 column
                        #df.replace() used for remove elements from entire table
df=df.fillna('')   #fillna use for fill empty cell and by using '' command we fill empty space with blank 
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```

```python
df =df.reset_index(drop=True) #we used drop=True bcoz it keep old index bcoz we saved all df previously
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('-','')
df["Phone_Number"]=df["Phone_Number"].str.replace('|','')
df["Phone_Number"]=df["Phone_Number"].str.replace('/','')
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:11])
df
```

```python
df=df.replace('nan--','')
df
```
